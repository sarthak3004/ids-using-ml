```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
col_names = ["duration","protocol_type","service","flag","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate","label","difficulty_level"]
```

```python
# importing dataset
train_data = pd.read_csv('KDDTrain+.txt',header=None, names=col_names)
train_data
```

```python
train_data.drop(['difficulty_level'],axis=1,inplace=True)
```

```python
train_data.label.value_counts()
```

```python
# changing attack labels to their respective attack class
def change_label(data):
  data.label.replace(['apache2','back','land','neptune','mailbomb','pod','processtable','smurf','teardrop','udpstorm','worm'],'Dos',inplace=True)
  data.label.replace(['ftp_write','guess_passwd','httptunnel','imap','multihop','named','phf','sendmail',
       'snmpgetattack','snmpguess','spy','warezclient','warezmaster','xlock','xsnoop'],'R2L',inplace=True)
  data.label.replace(['ipsweep','mscan','nmap','portsweep','saint','satan'],'Probe',inplace=True)
  data.label.replace(['buffer_overflow','loadmodule','perl','ps','rootkit','sqlattack','xterm'],'U2R',inplace=True)
```

```python
change_label(train_data)
train_data.label.value_counts()
```

# Normalisation

```python
# using standard scaler for normalizing
std_scaler = StandardScaler()
def normalization(df,col):
  for i in col:
    arr = df[i]
    arr = np.array(arr)
    df[i] = std_scaler.fit_transform(arr.reshape(len(arr),1))
  return df
```

```python
numeric_col_train = train_data.select_dtypes(include='number').columns
train_data = normalization(train_data.copy(),numeric_col_train)
train_data
```

# One Hot Encoding

```python
# selecting categorical data attributes
cat_col = ['protocol_type','service','flag']
# creating a dataframe with only categorical attributes
categorical_train = train_data[cat_col]
categorical_train.head()
```

```python
# one-hot-encoding categorical attributes using pandas.get_dummies() function
categorical_train = pd.get_dummies(categorical_train,columns=cat_col)
categorical_train.head()
```

# Binary Classification

```python
# changing attack labels into two categories 'normal' and 'abnormal'
bin_label_train = pd.DataFrame(train_data.label.map(lambda x:'normal' if x=='normal' else 'abnormal'))
```

```python
# creating a dataframe with binary labels (normal,abnormal)
bin_data_train = train_data.copy()
bin_data_train['label'] = bin_label_train
```

```python
# label encoding (0,1) binary labels (abnormal,normal)
le1 = preprocessing.LabelEncoder()
enc_label = bin_label_train.apply(le1.fit_transform)
bin_data_train['intrusion'] = enc_label
```

```python
le1.classes_
np.save("le1_classes.npy",le1.classes_,allow_pickle=True)
```

```python
# dataset with binary labels and label encoded column
bin_data_train.head()
```

```python
# one-hot-encoding attack label
bin_data_train = pd.get_dummies(bin_data_train,columns=['label'],prefix="",prefix_sep="") 
bin_data_train['label'] = bin_label_train
bin_data_train
```

```python
plt.figure(figsize=(5,5))
plt.pie(bin_data_train.label.value_counts(),labels=bin_data_train.label.unique(),autopct='%0.2f%%')
plt.title("Distribution of normal and abnormal labels")
plt.legend()
plt.show()
```

# Combining and feature extraction

```python
# creating a dataframe with only numeric attributes of binary class dataset and encoded label attribute 
numeric_bin = bin_data_train[numeric_col_train]
numeric_bin['intrusion'] = bin_data_train['intrusion']
```

```python
# finding the attributes which have more than 0.5 correlation with encoded attack label attribute 
corr= numeric_bin.corr()
corr_y = abs(corr['intrusion'])
highest_corr = corr_y[corr_y >0.5]
highest_corr.sort_values(ascending=True)
```

```python
# selecting attributes found by using pearson correlation coefficient
numeric_bin = bin_data_train[['count','srv_serror_rate','serror_rate','dst_host_serror_rate','dst_host_srv_serror_rate',
                         'logged_in','dst_host_same_srv_rate','dst_host_srv_count','same_srv_rate']]
```

```python
# selecting attributes found by using pearson correlation coefficient
numeric_bin = bin_data_train[['count','srv_serror_rate','serror_rate','dst_host_serror_rate','dst_host_srv_serror_rate',
                         'logged_in','dst_host_same_srv_rate','dst_host_srv_count','same_srv_rate']]
```

```python
# joining the selected attribute with the one-hot-encoded categorical dataframe
numeric_bin = numeric_bin.join(categorical_train)
# then joining encoded, one-hot-encoded, and original attack label attribute
bin_data_train = numeric_bin.join(bin_data_train[['intrusion','abnormal','normal','label']])
```

```python
# saving final dataset to disk
bin_data_train.to_csv("./datasets/bin_data.csv")
# final dataset for binary classification
bin_data_train
```

# Multi-class Classification

```python
# creating a dataframe with multi-class labels (Dos,Probe,R2L,U2R,normal)
multi_data = train_data.copy()
multi_label = pd.DataFrame(multi_data.label)
```

```python
# label encoding (0,1,2,3,4) multi-class labels (Dos,normal,Probe,R2L,U2R)
le2 = preprocessing.LabelEncoder()
enc_label = multi_label.apply(le2.fit_transform)
multi_data['intrusion'] = enc_label
```

```python
le2.classes_
np.save("le2_classes.npy",le2.classes_,allow_pickle=True)
```

```python
# one-hot-encoding attack label
multi_data = pd.get_dummies(multi_data,columns=['label'],prefix="",prefix_sep="") 
multi_data['label'] = multi_label
multi_data
```

```python
# pie chart distribution of multi-class labels
plt.figure(figsize=(10,10))
plt.pie(multi_data.label.value_counts(),labels=multi_data.label.unique(),autopct='%0.2f%%')
plt.title('Distribution of multi-class labels')
plt.legend()
plt.show()
```

# Combining and feature extraction

```python
# creating a dataframe with only numeric attributes of multi-class dataset and encoded label attribute 
numeric_multi = multi_data[numeric_col_train]
numeric_multi['intrusion'] = multi_data['intrusion']
```

```python
# finding the attributes which have more than 0.5 correlation with encoded attack label attribute 
corr = numeric_multi.corr()
corr_y = abs(corr['intrusion'])
highest_corr = corr_y[corr_y >0.5]
highest_corr.sort_values(ascending=True)
```

```python
# selecting attributes found by using pearson correlation coefficient
numeric_multi = multi_data[['count','logged_in','srv_serror_rate','serror_rate','dst_host_serror_rate',
                        'dst_host_same_srv_rate','dst_host_srv_serror_rate','dst_host_srv_count','same_srv_rate']]
```

```python
# joining the selected attribute with the one-hot-encoded categorical dataframe
numeric_multi = numeric_multi.join(categorical_train)
# then joining encoded, one-hot-encoded, and original attack label attribute
multi_data = numeric_multi.join(multi_data[['intrusion','Dos','Probe','R2L','U2R','normal','label']])
```

```python
# saving final dataset to disk
multi_data.to_csv('./datasets/multi_data.csv')

# final dataset for multi-class classification
multi_data
```
